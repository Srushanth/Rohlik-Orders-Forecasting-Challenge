```python
import pandas as pd
```

```python
df_train = pd.read_csv("../data/train.csv")
df_test = pd.read_csv("../data/test.csv")
```

```python
df = df_train[df_test.columns.to_list() + ["orders"]]
df.head()
```

```python
def pre_process(local_df: pd.DataFrame) -> pd.DataFrame:
    local_df = local_df.drop(columns=["date", "id"])
    local_df["holiday_name"] = local_df["holiday_name"].apply(lambda x: -1 if pd.isna(x) else 1)
    dummies = pd.get_dummies(local_df["warehouse"])
    local_df = pd.concat([local_df, dummies], axis=1)
    local_df = local_df.drop(columns=["warehouse"], axis=1)
    local_df = local_df.replace({True: 1, False: -1})
    return local_df
```

```python
df = df.drop(columns=["date", "id"])
```

```python
df["holiday_name"] = df["holiday_name"].apply(lambda x: -1 if pd.isna(x) else 1)
```

```python
dummies = pd.get_dummies(df["warehouse"])
```

```python
df = pd.concat([df, dummies], axis=1)
```

```python
df.drop(columns=["warehouse"], axis=1, inplace=True)
```

```python
df = df.replace({True: 1, False: -1})
```

```python
df
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
X = df.drop(columns=["orders"])
y = df["orders"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, train_size=0.2, random_state=42)
```

```python
regressor = DecisionTreeRegressor()
regressor.fit(X=x_train, y=y_train)
```

```python
y_pred = regressor.predict(x_test)
```

```python
mse = mean_squared_error(y_true=y_test, y_pred=y_pred)
mse
```

```python
r2 = r2_score(y_true=y_test, y_pred=y_pred)
r2
```

```python
pre_process(local_df=df_test)
```

```python
res = regressor.predict(pre_process(local_df=df_test))
```
